#### 2.1 Solving a ODE with TorchPhysics
Use TorchPhysics to solve the ODE for falling with a parachute:
\begin{align*}
    \partial_t^2 u(t) &= D(\partial_t u(t))^2 - g \\
    u(0) &= H \\
    \partial_t u(0) &= 0
\end{align*}
